# An example of putting a csv file (or pandas dataframe) into a netcdf file for many files based on example provided.

```python
# loading EASYMORE
from easymore.easymore import easymore

# initializing EASYMORE object
esmr = easymore()

# convert csv files of discharge, its flag and station info to netCDF
Flow = esmr.dataframe_to_netcdf_xr('./data/station_data/station_data.csv',
                                   data_frame_DateTime_column = 'Date_time',
                                   nc_file_name = 'station_FLOW.nc',
                                   nc_file_path = './output/',
                                   variable_name = 'Flow',
                                   unit_of_variable = 'm**3 s**-1',
                                   variable_long_name = 'daily discharge from HYDAT dataset',
                                   Fill_value = '-9999',
                                   station_info_data = './data/station_data/station_info.csv',
                                   station_info_column = 'info')
display(Flow)

Flag = esmr.dataframe_to_netcdf_xr('./data/station_data/station_data_flag.csv',
                                   data_frame_DateTime_column = 'Date_time',
                                   nc_file_name = 'station_FLAG.nc',
                                   nc_file_path = './output/',
                                   variable_name = 'Flag',
                                   unit_of_variable = '-',
                                   variable_long_name = 'Flags for discharge values',
                                   Fill_value = '-9999',
                                   station_info_data = './data/station_data/station_info.csv',
                                   station_info_column = 'info')
display(Flag)
```
